```
import pandas as pd
import matplotlib.pyplot as plt

# Data for K-Electric
kelectric_data = {
    "Customer Slab": ["Up to 100 Units", "101-200 Units", "201-300 Units", "301-400 Units", "401-500 Units", "501-600 Units", "601-700 Units", "Above 700 Units"],
    "K-Electric Cost (Rs/kWh)": [12.06, 14.38, 17.14, 19.53, 21.74, 23.16, 24.30, 28.95],
    "R.E. Source Cost (Rs/kWh)": [2.00, 2.00, 2.00, 2.00, 2.00, 2.00, 2.00, 2.00],
    "R.O.I. Period (Years)": [6.03, 7.19, 8.57, 9.77, 10.87, 11.58, 12.15, 14.47]
}

# Create DataFrame from K-Electric data
kelectric_df = pd.DataFrame(kelectric_data)

# Calculate the difference in cost per unit
kelectric_df['Cost Difference (Rs/kWh)'] = kelectric_df['K-Electric Cost (Rs/kWh)'] - kelectric_df['R.E. Source Cost (Rs/kWh)']

# Create a bar chart to compare the cost per unit
plt.figure(figsize=(10, 6))
plt.bar(kelectric_df['Customer Slab'], kelectric_df['Cost Difference (Rs/kWh)'], color=['blue', 'green'])
plt.xlabel('Customer Slab')
plt.ylabel('Cost Difference (Rs/kWh)')
plt.title('Comparison of Per Unit Cost between K-Electric and R.E. Source')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()

# Save the bar chart as a PNG image
plt.savefig('Comparison_of_Per_Unit_Cost.png')

# Display the DataFrame
print(kelectric_df)
```
