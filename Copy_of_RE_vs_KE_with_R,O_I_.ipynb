```

import matplotlib.pyplot as plt

# Define the data
customer_slabs = ["0-100 Units", "101-200 Units", "201-300 Units", "301-400 Units", "401-500 Units", "Above 500 Units"]
k_electric_cost = [8.95, 16.48, 24.01, 31.54, 39.07, 42.72]
r_e_source_cost = [5.00, 6.00, 7.00, 8.00, 9.00, 10.00]
roi_period = [5, 4, 3, 2, 1.5, 1]

# Create a combined figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Bar chart for cost comparison
width = 0.35
index = range(len(customer_slabs))
bars1 = axs[0].bar(index, k_electric_cost, width, label='K-Electric Cost', color='blue')
bars2 = axs[0].bar([i + width for i in index], r_e_source_cost, width, label='R.E. Source Cost', color='green')
axs[0].set_title('Comparison of Per Unit Cost')
axs[0].set_xlabel('Customer Slabs')
axs[0].set_ylabel('Cost per Unit (Rs.)')
axs[0].set_xticks([i + width / 2 for i in index])
axs[0].set_xticklabels(customer_slabs)
axs[0].legend()
axs[0].grid(True)

# Line chart for ROI period
axs[1].plot(customer_slabs, roi_period, marker='o', color='orange')
axs[1].set_title('ROI Period')
axs[1].set_xlabel('Customer Slabs')
axs[1].set_ylabel('ROI Period (Years)')
axs[1].grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
