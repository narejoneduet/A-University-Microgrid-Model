```python
import matplotlib.pyplot as plt
import numpy as np

# Create the data
years = np.arange(2023, 2027)
hydropower = [10470, 11270, 12070, 12870]
potential_hydropower = [20000, 22000, 24000, 26000]
actual_hydropower = [7500, 8500, 9500, 10500]
solar = [750, 850, 950, 1050]
potential_solar = [5000, 6000, 7000, 8000]
actual_solar = [350, 450, 550, 650]
wind = [300, 350, 400, 450]
potential_wind = [2000, 2200, 2400, 2600]
actual_wind = [100, 110, 120, 130]
biomass = [100, 110, 120, 130]
potential_biomass = [1000, 1100, 1200, 1300]
actual_biomass = [50, 55, 60, 65]
waste_to_energy = [50, 55, 60, 65]
potential_waste_to_energy = [700, 800, 900, 1000]
actual_waste_to_energy = [30, 35, 40, 45]

# Set the width of the bars
bar_width = 0.2

# Create the bar chart
plt.bar(years - bar_width * 2, hydropower, width=bar_width, label="Hydropower")
plt.bar(years - bar_width * 2, potential_hydropower, width=bar_width, label="Potential Hydropower", bottom=hydropower)
plt.bar(years - bar_width * 2, actual_hydropower, width=bar_width, label="Actual Hydropower", bottom=np.array(potential_hydropower) + np.array(hydropower))

plt.bar(years - bar_width, solar, width=bar_width, label="Solar")
plt.bar(years - bar_width, potential_solar, width=bar_width, label="Potential Solar", bottom=solar)
plt.bar(years - bar_width, actual_solar, width=bar_width, label="Actual Solar", bottom=np.array(potential_solar) + np.array(solar))

plt.bar(years, wind, width=bar_width, label="Wind")
plt.bar(years, potential_wind, width=bar_width, label="Potential Wind", bottom=wind)
plt.bar(years, actual_wind, width=bar_width, label="Actual Wind", bottom=np.array(potential_wind) + np.array(wind))

plt.bar(years + bar_width, biomass, width=bar_width, label="Biomass")
plt.bar(years + bar_width, potential_biomass, width=bar_width, label="Potential Biomass", bottom=biomass)
plt.bar(years + bar_width, actual_biomass, width=bar_width, label="Actual Biomass", bottom=np.array(potential_biomass) + np.array(biomass))

plt.bar(years + bar_width * 2, waste_to_energy, width=bar_width, label="Waste-to-Energy")
plt.bar(years + bar_width * 2, potential_waste_to_energy, width=bar_width, label="Potential Waste-to-Energy", bottom=waste_to_energy)
plt.bar(years + bar_width * 2, actual_waste_to_energy, width=bar_width, label="Actual Waste-to-Energy", bottom=np.array(potential_waste_to_energy) + np.array(waste_to_energy))

# Add a title and labels
plt.title("Prof. Dr. Ghous Narejo: Pakistan Renewable Energy Forecast")
plt.xlabel("Year")
plt.ylabel("KW Available")
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a list of region names
region_names = ['Africa', 'Asia', 'Europe', 'Latin America and the Caribbean', 'North America', 'Oceania']

# Create a list of potential values
potential_values = [1500, 12000, 3000, 2500, 3000, 100]

# Create a list of actual consumption values
actual_consumption_values = [100, 700, 1000, 500, 1500, 50]

# Set the width of the bars and the gap between groups of bars
bar_width = 0.4
bar_gap = 0.3

# Calculate the positions for the bars
x_pos_potential = range(len(region_names))
x_pos_actual = [pos + bar_width + bar_gap for pos in x_pos_potential]

# Create a bar chart
plt.bar(x_pos_potential, potential_values, width=bar_width, label="Potential")
plt.bar(x_pos_actual, actual_consumption_values, width=bar_width, label="Actual")

# Add a title to the chart
plt.title("Region-wise Potential vs Gap in RE Sources")

# Add labels to the x-axis and y-axis
plt.xlabel("Region")
plt.ylabel("TWh")

# Set x-axis ticks and labels
plt.xticks([pos + bar_width / 2 + bar_gap / 2 for pos in x_pos_potential], region_names)

# Add a legend to the chart
plt.legend()

# Adjust y-axis scale for better visualization
plt.ylim(0, 14000)

# Adjust figure size to provide more space for the bars
plt.figure(figsize=(10, 6))

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a list of region names in abbreviations
region_names = ['AFR', 'ASI', 'EUR', 'LAC', 'NAM', 'OCE']

# Create a list of potential values
potential_values = [1500, 12000, 3000, 2500, 3000, 100]

# Create a list of actual consumption values
actual_consumption_values = [100, 700, 1000, 500, 1500, 50]

# Set the width of the bars and the gap between regions
bar_width = 0.4
bar_gap = 0.6

# Calculate the positions for the bars
x_pos_potential = range(len(region_names))
x_pos_actual = [pos + bar_width + bar_gap for pos in x_pos_potential]

# Create a bar chart
plt.bar(x_pos_potential, potential_values, width=bar_width, label="Potential")
plt.bar(x_pos_actual, actual_consumption_values, width=bar_width, label="Actual")

# Add a title to the chart
plt.title("Region-wise Potential vs Gap in RE Sources")

# Add labels to the x-axis and y-axis
plt.xlabel("Region")
plt.ylabel("TWh")

# Set x-axis ticks and labels
plt.xticks([pos + (bar_width + bar_gap) / 2 for pos in x_pos_potential], region_names)

# Add the abbreviations below their respective bars
for i in range(len(region_names)):
    plt.text(x_pos_potential[i] + bar_width / 2, potential_values[i], region_names[i], ha='center', va='top', fontsize=10)

# Add a legend to the chart
plt.legend()

# Adjust y-axis scale for better visualization
plt.ylim(0, 14000)

# Show the chart
plt.show()
```
