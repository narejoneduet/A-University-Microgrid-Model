````
import matplotlib.pyplot as plt

# Energy mix data (percentage values)
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
coal = [33.2, 35.2, 37.1, 39.5, 39.8, 39.4, 40.1, 40.3, 40.9, 41.5]
gas = [34.7, 33.6, 32.1, 31.5, 31.2, 30.8, 30.3, 30.0, 29.5, 29.1]
hydro = [29.1, 28.6, 28.7, 28.0, 27.6, 28.0, 26.5, 26.1, 25.5, 24.8]
nuclear = [2.1, 2.6, 2.9, 3.0, 2.9, 2.8, 2.9, 2.9, 2.9, 3.0]
renewable = [0.9, 0.9, 0.9, 0.9, 0.9, 0.9, 0.9, 0.9, 0.9, 1.6]

# Plotting the energy mix
plt.figure(figsize=(10, 6))
plt.bar(years, coal, label='Coal', width=0.6)
plt.bar(years, gas, bottom=coal, label='Gas', width=0.6)
plt.bar(years, hydro, bottom=[coal[j] + gas[j] for j in range(len(coal))], label='Hydro', width=0.6)
plt.bar(years, nuclear, bottom=[coal[j] + gas[j] + hydro[j] for j in range(len(coal))], label='Nuclear', width=0.6)
plt.bar(years, renewable, bottom=[coal[j] + gas[j] + hydro[j] + nuclear[j] for j in range(len(coal))],
        label='Renewable', width=0.6)

# Customizing the plot
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Prof. Dr. Ghous Narejo: Pakistan Energy Mix (2013-2022)')
plt.xticks(years)
plt.yticks(range(0, 101, 10))
plt.legend(loc='upper left')

# Displaying the plot
plt.show()
````
