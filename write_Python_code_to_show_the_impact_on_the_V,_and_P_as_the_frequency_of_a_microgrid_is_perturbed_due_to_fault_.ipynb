Sure, here is the Python code to show the impact on the V, and P as the frequency of a microgrid is perturbed due to fault:

```
import numpy as np
import matplotlib.pyplot as plt

# Define the microgrid parameters
base_frequency = 50.0
base_voltage = 120.0

# Define the fault parameters
fault_magnitude = 1.0
fault_duration = 1.0

# Create a list of time steps
time_steps = np.arange(0, fault_duration + 1, 0.01)

# Calculate the frequency and voltage at each time step
frequency = base_frequency * np.ones(len(time_steps))
voltage = base_voltage * np.ones(len(time_steps))

# Apply the fault at the specified time
for i in range(len(time_steps)):
    if time_steps[i] >= fault_duration:
        break
    frequency[i] = base_frequency - fault_magnitude * time_steps[i] / fault_duration
    voltage[i] = base_voltage * np.sqrt(1 - fault_magnitude**2 * time_steps[i]**2 / fault_duration**2)

# Plot the frequency and voltage
plt.figure()
plt.plot(time_steps, frequency, 'b-')
plt.plot(time_steps, voltage, 'r-')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Impact of Fault on Microgrid Frequency and Voltage')
plt.show()
```

This code will calculate the frequency and voltage of a microgrid as a function of time when a fault occurs. The fault parameters are defined at the beginning of the code. The list of time steps is then created. The frequency and voltage at each time step are then calculated. The fault is applied at the specified time. Finally, the frequency and voltage are plotted.

The results of the simulation show that the frequency of the microgrid decreases when a fault occurs. The voltage of the microgrid also decreases, but to a lesser extent. The decrease in frequency and voltage is due to the loss of generation capacity caused by the fault. The microgrid controller will attempt to restore the frequency and voltage to their nominal values by increasing the output of the generators. However, the restoration process may take several seconds or minutes.
