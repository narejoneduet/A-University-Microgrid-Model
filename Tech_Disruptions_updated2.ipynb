```
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import re
from datetime import datetime

# Define the tech disruptions and their respective timeframes
tech_disruptions = [
    ("Personal Computer Revolution", "1970s", "1980s"),
    ("Internet", "1990s", ""),
    ("E-commerce Boom", "late 1990s", "early 2000s"),
    ("Mobile Revolution", "2000s", ""),
    ("Social Media", "mid-2000s", ""),
    ("Ride-Sharing", "2010s", ""),
    ("Streaming Services", "2010s", ""),
    ("Electric Vehicles", "2010s", ""),
]

# Prepare the data for plotting
start_dates = []
end_dates = []
labels = []

for label, start, end in tech_disruptions:
    start_year = re.findall(r"\d{4}", start)
    start_date = datetime.strptime(start_year[0], "%Y") if start_year else None

    end_year = re.findall(r"\d{4}", end)
    end_date = datetime.strptime(end_year[0], "%Y") if end_year else datetime.now()

    start_dates.append(start_date)
    end_dates.append(end_date)
    labels.append(label)

# Plot the timeline
fig, ax = plt.subplots()
ax.barh(labels, end_dates, left=start_dates)
ax.xaxis_date()

# Customize the plot
ax.set_xlabel("Timeline")
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.yaxis.tick_left()
ax.invert_yaxis()

plt.title("Tech Disruptions Timeline")
plt.tight_layout()
plt.show()
```
